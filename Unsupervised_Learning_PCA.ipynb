# Supervised Learning

Supervised learning, also known as supervised machine learning, is a subcategory of machine learning and artificial intelligence. It is defined by its use of labeled datasets to train algorithms that to classify data or predict outcomes accurately. As input data is fed into the model, it adjusts its weights until the model has been fitted appropriately, which occurs as part of the cross validation process. Supervised learning helps organizations solve for a variety of real-world problems at scale, such as classifying spam in a separate folder from your inbox.

image.png

An example of a supervised learning algorithm is neural networks. Primarily leveraged for deep learning algorithms, neural networks process training data by mimicking the interconnectivity of the human brain through layers of nodes. Each node is made up of inputs, weights, a bias (or threshold), and an output. If that output value exceeds a given threshold, it fires or activates the node, passing data to the next layer in the network. Neural networks learn this mapping function through supervised learning, adjusting based on the loss function through the process of gradient descent. When the cost function is at or near zero, we can be confident in the models accuracy to yield the correct answer.

# Unsupervised Learning

Unsupervised machine learning uses machine learning algorithms to analyze and cluster unlabeled datasets. These algorithms discover hidden patterns or data groupings without the need for human intervention. These models are used for three main tasks--clustering, association, and dimensionality reduction.

## Clustering

Clustering is a technique which groups unlabeled data based on similarities or differences. Clustering algorithms can be used to process data into groups represented by patterns in information. These algorithms can be categorized into a few types, specifically exclusive, overlapping, hierarchical, and probabilistic.

Gaussian Mixture Models are a form of clustering. As you learned in the previous learning module they calculate which subpopulation data belongs to within an overall population. These subpopulations are clusters.

## Association

An association rule is a rule-based method for finding relationships between variables in a given dataset. These methods are frequently used for market basket analysis, allowing companies to better understand relationships between different products. Understanding consumption habits of customers enables businesses to develop better cross-selling strategies and recommendation engines. Examples of this can be seen in Amazons Customers Who Bought This Item Also Bought or Spotifys "Discover Weekly" playlist. While there are a few different algorithms used to generate association rules, such as Apriori, Eclat, and FP-Growth, the Apriori algorithm is most widely used.

image.png

Apriori algorithms have been popularized through market basket analyses, leading to different recommendation engines for music platforms and online retailers. They are used within transactional datasets to identify frequent itemsets, or collections of items, to identify the likelihood of consuming a product given the consumption of another product. For example, if I play Black Sabbaths radio on Spotify, starting with their song Orchid, one of the other songs on this channel will likely be a Led Zeppelin song, such as Over the Hills and Far Away. This is based on my prior listening habits as well as the ones of others. Apriori algorithms use a hash tree to count itemsets, navigating through the dataset in a breadth-first manner.

## Dimensionality Reduction

While more data generally yields more accurate results, it can also impact the performance of machine learning algorithms (e.g. overfitting) and it can also make it difficult to visualize datasets. Dimensionality reduction is a technique used when the number of features, or dimensions, in a given dataset is too high. It reduces the number of data inputs to a manageable size while also preserving the integrity of the dataset as much as possible. It is commonly used in the preprocessing data stage, and there are a few different dimensionality reduction methods that can be used.

One example of dimensionality reduction is principal component analysis (PCA) is a type of dimensionality reduction algorithm which is used to reduce redundancies and to compress datasets through feature extraction. This method uses a linear transformation to create a new data representation, yielding a set of "principal components." The first principal component is the direction which maximizes the variance of the dataset. While the second principal component also finds the maximum variance in the data, it is completely uncorrelated to the first principal component, yielding a direction that is perpendicular, or orthogonal, to the first component. This process repeats based on the number of dimensions, where a next principal component is the direction orthogonal to the prior components with the most variance.

## Clustering

 In this example, we demonstrate the K-Means clustering algorithm. We start by generating synthetic data using the make_blobs function, which creates clusters of data points. Then, we apply the K-Means algorithm using the KMeans class from the scikit-learn library. Finally, we visualize the clusters and centroids using matplotlib. K-Means is commonly used in various applications such as customer segmentation, image compression, and anomaly detection.

```python
# Import necessary libraries
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

# Visualize the clusters
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75)
plt.title('K-Means Clustering')
plt.show()
```

## Association

In this example, we illustrate how to apply the Apriori algorithm using the mlxtend library. We create a sample transaction dataset and convert it into a one-hot encoded format. This means representing each transaction as a binary vector where each element corresponds to an item, and a value of 1 indicates the presence of the item in the transaction. Then, we use the Apriori algorithm to find frequent itemsets with a minimum support threshold. Finally, we generate association rules based on these frequent itemsets and evaluate them using metrics such as lift.

In the resulting table there are several columns that contribute to showing the results of the algorithm:

1. Antecedent: This column lists the antecedent or premise of each association rule. The antecedent is a set of items that are found together in transactions and serve as the basis for making predictions about the consequent.

2. Consequent: This column lists the consequent or conclusion of each association rule. The consequent is the item or items that are likely to occur in transactions given the presence of the antecedent.

3. Support: Support quantifies the frequency with which a particular combination of items (antecedent and consequent together) appears in the dataset. It is calculated as the proportion of transactions that contain both the antecedent and the consequent.

4. Confidence: Confidence measures the reliability or strength of an association rule. It indicates the conditional probability of observing the consequent in a transaction given that the antecedent is present. It is calculated as the proportion of transactions containing both the antecedent and the consequent out of all transactions containing the antecedent.

5. Lift: Lift is a measure of how much more likely the consequent is to occur in transactions containing the antecedent compared to its expected occurrence by chance. It is calculated as the ratio of the observed support to the expected support under independence. A lift value greater than 1 indicates that the antecedent and consequent occur together more often than expected, suggesting a positive association.

```python
# Install and import necessary libraries
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# Create sample transaction dataset
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]

# Convert dataset to one-hot encoded format
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(df, min_support=0.5, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Display association rules
print("Association:")
print(rules)
```

## Dimensionality Reduction

In this example, we demonstrate how to apply PCA to the Iris dataset, which contains measurements of iris flowers. We project the data onto a lower-dimensional space defined by the principal components and visualize the reduced dimensional data using matplotlib. This reduces the dimensionality of the dataset while retaining as much variance as possible. We visualize the reduced-dimensional data using a scatter plot, where each point represents an observation (iris flower) in the reduced space. This visualization helps us understand the underlying structure of the data and potentially identify clusters or patterns. PCA is valuable for data visualization, feature extraction, and noise reduction in high-dimensional datasets.

```python
# Import necessary libraries
from sklearn.decomposition import PCA

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualize reduced dimensional data
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.title('PCA for Iris Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
